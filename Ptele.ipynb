```python
customer_df = customers.merge(network_usage, on="customer_id", how="left")
customer_df = customer_df.merge(network_issues, on="customer_id", how="left")
customer_df = customer_df.merge(billing, on="customer_id", how="left")
```

```python
issue_summary = network_issues.groupby("customer_id").size().reset_index(name="issue_count")
```

```python
issue_summary.head()
```

```python
customer_df = customer_df.merge(issue_summary, on="customer_id", how="left")
```

```python
customer_df["issue_count"] = customer_df["issue_count"].fillna(0)
```

```python
customer_df[["customer_id", "issue_count"]].head()
```

```python
def calculate_risk(row):
    score = 0
    if row["data_used_gb"] > 10:
        score += 3
    if row["issue_count"] > 5:
        score += 3
    if row["payment_status"] == "delayed":
        score += 4
    if row["data_used_gb"] < 5:
        score += 1
    return score
```

```python
customer_df["risk_score"] = customer_df.apply(calculate_risk, axis=1)
```

```python
customer_df[["customer_id", "risk_score"]].head()
```

```python
customer_df[["customer_id", "risk_score","data_used_gb","payment_status"]]
```

```python
def risk_category(score):
    if score >= 6:
        return "High Risk"
    elif score >= 4:
        return "Medium Risk"
    else:
        return "Low Risk"
customer_df["risk_level"] = customer_df["risk_score"].apply(risk_category)
```

```python
customer_df["risk_level"].value_counts()
```

```python
def calculate_risk(row):
    score = 0
    
    if row.get("data_used_gb", 0) > 20:
        score += 3
    
    if row.get("issue_count", 0) > 5:
        score += 3
    
    if row.get("payment_status") == "delayed":
        score += 4
    
    if row.get("data_used_gb", 0) < 2:
        score += 1
        
    return score
```

```python
customer_df["risk_score"] = customer_df.apply(calculate_risk, axis=1)
```
